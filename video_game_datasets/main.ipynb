```python
import pandas
import numpy
```

```python
df = pandas.read_csv("Video_Games.csv")
```

```python
df.head()
```

```python
df.drop("index", axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
# Sadece sayısal sütunları seçme
numeric_columns = df.select_dtypes(include='number').columns

# Sayısal sütunları ortalama değer ile doldurma
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
# ÖNEMLİİİİ
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
degiscek_sutunlar = ["User_Count","Critic_Count","Critic_Score","Global_Sales","Other_Sales","JP_Sales","EU_Sales","NA_Sales","Year_of_Release"]
```

```python
df[degiscek_sutunlar] = df[degiscek_sutunlar].astype("int32")
```
